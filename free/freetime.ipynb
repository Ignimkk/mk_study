```python
import matplotlib.pyplot as plt
from matplotlib import font_manager

def check_korean_font(font_name):
    try:
        # Create a figure and axis
        fig, ax = plt.subplots()

        # Set the font for the axis labels
        ax.set_xlabel('한글 텍스트', fontname=font_name, fontsize=14)
        ax.set_ylabel('한글 텍스트', fontname=font_name, fontsize=14)

        # Set the title with Korean text
        ax.set_title('한글 폰트 테스트', fontname=font_name, fontsize=16)

        # Display the plot
        plt.show()

    except Exception as e:
        print(f"Error: {e}")
        print(f"Make sure the Korean font '{font_name}' is installed on your system.")

# Replace 'YourKoreanFont' with the actual name of the Korean font you want to check
korean_font_name = 'YourKoreanFont'

check_korean_font(korean_font_name)
```

```python
from PIL import Image
import pytesseract


image_path = '/home/addinedu/Pictures/Screenshots/Screenshot from 2024-02-28 09-40-01.png'

image = Image.open(image_path)

gray_image = image.convert('L')
contrast_factor = -10


adjusted_image = gray_image.point(lambda x: x + contrast_factor if x + contrast_factor <= 255 else 255)

threshold = 95 #90 
binary_image = adjusted_image.point(lambda x: 255 if x < threshold else 0)

text = pytesseract.image_to_string(binary_image, lang='eng+kor')
binary_image.show(title='Original Image')

print("text :")
print(text)
```

```python
from PIL import Image, ImageFilter
import pytesseract

image_path = '/home/addinedu/Pictures/Screenshots/Screenshot from 2024-02-27 17-45-44.png'

image = Image.open(image_path)

gray_image = image.convert('L')

sharpened_image = gray_image.filter(ImageFilter.UnsharpMask(radius=2, percent=100, threshold=0))


contrast_factor = -10
adjusted_image = sharpened_image.point(lambda x: x + contrast_factor if x + contrast_factor <= 255 else 255)

threshold = 95
binary_image = adjusted_image.point(lambda x: 255 if x < threshold else 0)

text = pytesseract.image_to_string(binary_image, lang='eng+kor')
binary_image.show(title='Original Image')

print("text :")
print(text)
```
