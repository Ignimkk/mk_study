```python
import cv2
import torch

# Load YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'custom', path ='/home/addinedu/yolo/yolov5/runs/train/rock_paper_yolo5/weights/best.pt')

# Open webcam
cap = cv2.VideoCapture(2)

while cap.isOpened():
    # Read frame from webcam
    ret, frame = cap.read()
    if not ret:
        break
    results = model(frame)

    # Process inference results
    predictions = results.xyxy[0].cpu().numpy()

    # Draw bounding boxes around detected objects
    for pred in predictions:
        x1, y1, x2, y2, conf, class_id = pred
        if conf >= 0.7:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
            cv2.putText(frame, f'Class: {int(class_id)}, Conf: {conf:.2f}', (int(x1), int(y1) - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
            center_x = (x1 + x2) // 2
            center_y = (y1 + y2) // 2
            print(f"Center Coordinates: ({center_x}, {center_y})")
    # Display the result
    cv2.imshow('YOLOv5 Object Detection', frame)

    # Check for 'q' key press to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release video capture and close all windows
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import torch
import numpy as np

def nothing(x):
    pass

# Create a black window
cv2.namedWindow('Adjust Parameters')

# Create trackbars for adjusting parameters
cv2.createTrackbar('Brightness', 'Adjust Parameters', 0, 100, nothing)
cv2.createTrackbar('Contrast', 'Adjust Parameters', 0, 100, nothing)
cv2.createTrackbar('Exposure', 'Adjust Parameters', 0, 100, nothing)

# Load YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/home/addinedu/yolo/yolov5/runs/train/cigar_yolov5s_2/weights/best.pt')

# Open webcam
cap = cv2.VideoCapture(0)

while cap.isOpened():
    # Read frame from webcam
    ret, frame = cap.read()
    
    if not ret:
        break
    
    

    brightness = cv2.getTrackbarPos('Brightness', 'Adjust Parameters') / 50.0
    contrast = cv2.getTrackbarPos('Contrast', 'Adjust Parameters') / 50.0
    exposure = cv2.getTrackbarPos('Exposure', 'Adjust Parameters') / 50.0

    # Adjust brightness, contrast, and exposure
    adjusted = np.clip(frame * contrast + brightness * 255, 0, 255).astype(np.uint8)
    adjusted = cv2.convertScaleAbs(adjusted, alpha=exposure)

    results = model(adjusted)
    # Process inference results
    predictions = results.xyxy[0].cpu().numpy()

    # Draw bounding boxes around detected objects with confidence greater than or equal to 0.6
    for pred in predictions:
        x1, y1, x2, y2, conf, class_id = pred
        if conf >= 0.6:
            cv2.rectangle(adjusted, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
            cv2.putText(adjusted, f'Class: {int(class_id)}, Conf: {conf:.2f}', (int(x1), int(y1) - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
            center_x = (x1 + x2) // 2
            center_y = (y1 + y2) // 2
            print(f"Center Coordinates: ({center_x}, {center_y})")
            
    # Display the result
    cv2.imshow('YOLOv5 Object Detection', adjusted)

    # Check for 'q' key press to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release video capture and close all windows
cap.release()
cv2.destroyAllWindows()
```
