```python
import os
from PIL import Image
import pytesseract

def process_image(image_path):
    # 이미지 로드
    img = Image.open(image_path)

    # 이미지를 흑백으로 변환
    img_gray = img.convert('L')

    # 이진화를 위한 임계값 적용
    threshold = 180  # 이미지에 따라 이 임계값을 조정할 수 있습니다
    img_bin = img_gray.point(lambda p: p > threshold and 255)

    # 전처리된 이미지 저장
    img_bin.save(image_path + '_binarized.jpg')

    # Tesseract를 사용하여 OCR 수행
    text = pytesseract.image_to_string(img_bin, config='--oem 3')
    return text

def process_directory(directory_path):
    # 지정된 디렉토리의 모든 이미지 처리
    for filename in os.listdir(directory_path):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp')):
            full_path = os.path.join(directory_path, filename)
            print(f"{full_path} 처리 중...")
            text = process_image(full_path)
            print(f"{filename}에서 감지된 텍스트:")
            print(text)
            print("-" * 80)

if __name__ == "__main__":
    images_directory = '/home/mk/captue for ocr/4'
    process_directory(images_directory)
```
