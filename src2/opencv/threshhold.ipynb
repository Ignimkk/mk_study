```python
img = cv2.imread('/home/addinedu/dev_ws/opencv/data/crossword.jpg', 0)
print(f'type(img) = {type(img)}, img.shape = {img.shape}')
plt.imshow(img, cmap='gray');
```

```python
def show_pic(img):
    fig = plt.figure(figsize = (15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_pic(img)
```

```python
ret, thresh1 = cv2.threshold(img,80,255, cv2.THRESH_BINARY)
show_pic(thresh1)
```

```python
thresh2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 10)
show_pic(thresh2)
```

```python
show_pic(thresh1)
```

```python
blended = cv2.addWeighted(src1= thresh1, alpha= 0.6, src2= thresh2, beta=0.4, gamma=0)
show_pic(blended)
```

```python
thresh2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 8)
show_pic(thresh2)
```

```python
blended = cv2.addWeighted(src1= thresh1, alpha= 0.6, src2= thresh2, beta=0.4, gamma=0)
show_pic(blended)
```
