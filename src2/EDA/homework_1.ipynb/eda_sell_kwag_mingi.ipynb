```python
def scrape_data(driver):


    req = driver.page_source
    soup = BeautifulSoup(req, 'html.parser')

    try:
        house = soup.find_all('a', attrs={'href': '#a'})
        worked = soup.find_all('td', attrs={'class': 'word_cut'})

        for i in range(len(house)):
            house_name = house[i+2].get_text()
            house_name_list.append(house_name)

            qq = 2*i+1
            rr = worked[qq].get_text()
            worked_list.append(rr)

            yy = soup.find_all('tr', attrs={'data-hsecd':'02'})
            nn = yy[i]
            pp = nn.find('td', attrs={'class':'txt_l'}).get_text()
            tt = nn.find_all('td')

            for o in range(len(tt)):
                if o == 4:
                    mozip_day= tt[o].get_text()
                    mozip_day_list.append(mozip_day)
                elif o == 5:
                    blue_day = tt[o].get_text()
                    blue_day_list.append(blue_day)
                elif o == 6:
                    lucky_day = tt[o].get_text()
                    lucky_day_list.append(lucky_day)
                else:
                    print(i)
        combined_list = list(zip(house_name_list, worked_list, mozip_day_list, blue_day_list, lucky_day_list ))
        return combined_list
    
    except Exception as e:
        print(f"gg: {e}")

        return None
```

```python
def move_to_page(driver, page_num):
    try:
        page_move = driver.find_element(By.ID, "paging")
        page_move_list = page_move.find_elements(By.TAG_NAME, 'a')

        if page_num <= len(page_move_list):
            page_2dong = driver.find_element(By.XPATH, f'//*[@id="paging"]/a[{page_num}]')
            try:
                page_2dong.click()
            except:
                print("z")
            time.sleep(2)
        else:
            print(f"Page number {page_num}.")
    except Exception as e:
        print(f"Error: {e}")
```

```python

house_name_list = []
worked_list = []
mozip_day_list = []
blue_day_list = []
lucky_day_list = []
all_data = []
combined_list = []
page_move = driver.find_element(By.ID, "paging")
page_move_list = page_move.find_elements(By.TAG_NAME, 'a')
total_pages = len(page_move_list)

for page_number in range(1, total_pages + 1):  # Adjust 'total_pages' accordingly
    move_to_page(driver, page_number)
    
    req = driver.page_source
    soup = BeautifulSoup(req, 'html.parser')

    scraped_data = scrape_data(driver)

    if scraped_data:
        all_data.extend(scraped_data)
  
    
all_data
```
