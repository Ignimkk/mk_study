```python
import folium 

m = folium.Map(location=[45.5236, -122.6750]) # 지도를 만들건데, 내가 보고 싶은 지도의 중심은 location을 통해서 위도경도를 지정한다.
m
```

```python
Titles_Builtin = ["OpenStreetMap", "Cartodb Positron", "Cartodb dark_matter"]   # 지도의 배경 스타일을 지정한다
folium.Map(location=[37.4778772, 126.8806482], tiles=Titles_Builtin[0], zoom_start=13) #지도의 중점과 배경스타일, 확대 스케일을 지정함.
```

```python
MyHome = [37.4778772, 126.8806482]
MyNeighbor = [37.47791, 126.8791]

my_map = folium.Map(location=MyHome, tiles=Titles_Builtin[0], zoom_start=18)

folium.Marker(MyHome, popup="<i>My Sweey Home</i>").add_to(my_map) #my_map에 내가 표시하고 싶은 곳을 마커를 통해서 지정한다.
folium.Marker(MyNeighbor, popup="<i>My Neighbor</i>").add_to(my_map)
```

```python
my_map
```

```python
MyFriend = [37.47782, 126.88062]

folium.Marker(
    location = MyHome,
    popup = "My Sweet Home",
    icon = folium.Icon(color="green"),
).add_to(my_map)

folium.Marker(
    location = MyNeighbor,
    popup = "My Neighbor",
    icon = folium.Icon(icon="cloud"),
).add_to(my_map)

folium.Marker(
    location = MyFriend,
    popup = "My Friend",
    icon = folium.Icon(icon="info-sign"),
).add_to(my_map)

my_map
```

```python
folium.Circle(
    radius = 100,
    location = MyHome,
    popup = "My Sweet Home",
    color = "crimson",
    fill = False
).add_to(my_map)

folium.CircleMarker(
    radius = 50,
    location = MyNeighbor,
    popup = "My Neighbor",
    color = "#3186cc",
    fill = True,
    fill_color = "#3186cc" #구글에서 색상코드 찾아보기
).add_to(my_map)
```

```python
my_map.add_child(folium.LatLngPopup()) #지도를 클릭하면 위도와 경도를 표시함.
my_map
```
