```python
menu = driver.find_element(By.ID, "main_navbar")
li_list = menu.find_elements(By.TAG_NAME, "li")
for idx, li in enumerate(li_list):
    print("[" + str(idx) + "] ====================")
    print(li.get_attribute('innerHTML'))
```

```python
about = li_list[0]
about.click()
link_list = about.find_elements(By.CLASS_NAME, "dropdown-item")
for link in link_list:
    print(link.text)
```

```python
history = about.find_element(By.XPATH, '''//*[@id="main_navbar"]/ul/li[1]/div/a[5]''' )
history.click()
```

```python
driver.back()
```

```python
from selenium.webdriver import ActionChains
import time

actions = ActionChains(driver)
actions.click(about)
time.sleep(1)
actions.click(history)
time.sleep(1)
actions.perform()
```
